## Data Processing (40 points)

The task is to create bag-of-words features: tokenize the text, index each token, represent the sentence as a dictionary of tokens and their counts, limit the vocabulary to $n$ most frequent tokens. In the lab we use built-in `sklearn` function, `sklearn.feature_extraction.text.CountVectorizer`. 
**In this HW, you are required to implement the `Vectorizer` on your own without using `sklearn` built-in functions.**

Function `preprocess_data` takes the list of texts and returns list of (lists of tokens). 
You may use [spacy](https://spacy.io/) or [nltk](https://www.nltk.org/) text processing libraries in `preprocess_data` function. 

Class `Vectorizer` is used to vectorize the text and to create a matrix of features.


```python
import re
import nltk
# nltk.download('stopwords')
from nltk import word_tokenize
from nltk.stem import LancasterStemmer


def preprocess_data(data):
    # This function should return a list of lists of preprocessed tokens for each message
    """
    YOUR CODE GOES HERE
    """
   
    # removing all words except a-z
    data = data.apply(lambda x: re.sub("[^a-zA-Z]", " ", x))

    # tokenize
    word_token = [word_tokenize(text) for text in data]

    # removing stopwords and including words that have length of more than 2
    # setting stopwords and list variable
    stopwords = nltk.corpus.stopwords.words('english')
    all_tokens = []
    
    # for each list of lists, literate to see single list and add temp list variable(filtered_words)
    for sentence in word_token:
        filtered_words = []
        
        # for single list, iterate by words
        # lower every words, and append when given word is not a stopword and its length >= 2
        # length constraint is used since CountVectorizer's default word length is >= 2
        for word in sentence:
            word = word.lower()

            if word not in stopwords and len(word) >= 2:
                filtered_words.append(word)
        
        # finally, add temp list into 'all_token' list
        all_tokens.append(filtered_words)


    # stemming by using LancasterStemmer
    stemmer = LancasterStemmer()
    all_tokens = pd.Series(all_tokens).apply(lambda x: [stemmer.stem(y) for y in x])
    
    # Now, everything is done. Just convert into list and return
    preprocessed_data = list(all_tokens)
    return preprocessed_data

train_data = preprocess_data(train_texts)
val_data = preprocess_data(val_texts)
test_data = preprocess_data(test_texts)
```

```python
import numpy as np
import collections
from scipy.sparse import csr_matrix

class Vectorizer():
    def __init__(self, max_features):
        self.max_features = max_features
        self.vocab_list = None
        self.token_to_index = None

    def fit(self, dataset):
        # Create a vocab list, self.vocab_list, using the most frequent "max_features" tokens
        # Create a token indexer, self.token_to_index, that will map each token in self.vocab 
        # to its corresponding index in self.vocab_list
        """
        YOUR CODE GOES HERE
        """
        
        # Creating vocal_list with max_features
        words = collections.Counter(word for sentence in dataset for word in sentence)
        words_sorted = (dict(sorted(words.items(), key = lambda x: -x[1])))
        self.vocab_list = list(words_sorted.keys())[:self.max_features]
        
        # Creating a token indexer
        vocab = {}
        for index, word in enumerate(self.vocab_list):
            vocab[word] = index
        self.token_to_index = vocab
        
    def transform(self, dataset):
        # This function transforms text dataset into a matrix, data_matrix
        """
        YOUR CODE GOES HERE
        """
        # initializing data matrix and needed csr_matrix
        data_matrix = np.zeros((len(dataset), len(self.vocab_list)))
        row, col, val = [], [], []
        
        # iterate each dataset
        for idx, sentence in enumerate(dataset):
            
            # for each sencence, create a dictionary with frequency by using collections.Counter function
            count_word = dict(collections.Counter(sentence))
            
            # for given counter variable, iterate by each word to check if the word is in fitted word libraray
            for word, count in count_word.items():
                
                # if so, find index number of library 
                if word in self.vocab_list:
                    col_index = self.token_to_index[word]
                    
                    # this if statement is just to make sure that col_index has a value
                    # if so, index for (row, col) with value(freq) 
                    if col_index >= 0:
                        row.append(idx)
                        col.append(col_index)
                        val.append(count)
        
        # now make into csr_matrix and convert into array
        data_matrix = csr_matrix((val, (row,col)), shape = (len(dataset), len(self.vocab_list))).toarray()
        
        return data_matrix
```

```python
def accuracy_score(y_true, y_pred): 
    # Calculate accuracy of the model's prediction
    """
    YOUR CODE GOES HERE
    """    
    # accuray can be calculated by count(y_true = y_pred) / total number
    accuracy = np.mean(y_true == y_pred)
    return accuracy

def f1_score(y_true, y_pred): 
    # Calculate F1 score of the model's prediction
    """
    YOUR CODE GOES HERE
    """
    
    # recall that f1 score = 2 * (precision * recall) / (precision + recall)
    # precision = tp / (tp + fp)
    # recall = tp / (tp + fn)
    tp, tn, fp, fn = 0, 0, 0, 0
    precision, recall = 0, 0
    
    for i in range(len(y_true)):
        if y_true[i] == 1 and y_pred[i] == 1: tp += 1
        elif y_true[i] == 0 and y_pred[i] == 0: tn += 1
        elif y_true[i] == 0 and y_pred[i] == 1: fp += 1
        elif y_true[i] == 1 and y_pred[i] == 0: fn += 1            
    
    precision = tp / (tp + fp)
    recall = tp / (tp + fn)
        
    f1 = 2 * (precision * recall) / (precision + recall)
    return f1
```

**Question.**
Is accuracy the metric that logistic regression optimizes while training? If no, which metric is optimized in logistic regression?

**Your answer:** 
- No. Logistic regression optimizes the loss function of the likelihood(probability) of the correct answer. By formula, it minimizes the negative log of the likelihood (probability) of the correct answer (= maximizing the likelihood of the correct answer)  

- While accuracy is not the metric that the logistic regression optimizes, it is one of the model evaluation method. Since accuracy is calculated by comparing predicted result and actual result to see if those are same or not, it is related to the likelihood(probability) of the correct answer.
