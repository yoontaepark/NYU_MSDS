```python
# Q19 gorns
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
import matplotlib.pyplot as plt

gorns = pd.read_csv('./gorns.csv', header = None, names = ['col1', 'col2'])
gorns.head(3)
```

```python
test = np.array(gorns)
```

```python
from sklearn.metrics import pairwise_distances_argmin

def KMeans(X, n_clusters, rseed=2):
    # 1. Randomly choose clusters
    rng = np.random.RandomState(rseed)
    i = rng.permutation(X.shape[0])[:n_clusters]
#     print(i.shape)
    centers = X[i]
    
    while True:
        # 2a. Assign labels based on closest center
        labels = pairwise_distances_argmin(X, centers)
        
        # 2b. Find new centers from means of points
        new_centers = np.array([X[labels == i].mean(0)
                                for i in range(n_clusters)])
        
        # 2c. Check for convergence
        if np.all(centers == new_centers):
            break
        centers = new_centers
    
    return centers, labels
```

```python
centers, labels = KMeans(test, 5)
plt.scatter(test[:, 0], test[:, 1], c=labels,
            s=50, cmap='viridis');
```

```python
c_0 = gorns[gorns['cluster'] == 0]
c_1 = gorns[gorns['cluster'] == 1]
c_2 = gorns[gorns['cluster'] == 2]
c_3 = gorns[gorns['cluster'] == 3]
c_4 = gorns[gorns['cluster'] == 4]
```

```python
c_0_dist = sum(pow((c_0['col1'] - centers[0][0]), 2) + pow(c_0['col2'] - centers[0][1], 2))
c_1_dist = sum(pow(c_1['col1'] - centers[1][0], 2) + pow(c_1['col2'] - centers[1][1], 2))
c_2_dist = sum(pow(c_2['col1'] - centers[2][0], 2) + pow(c_2['col2'] - centers[2][1], 2))
c_3_dist = sum(pow(c_3['col1'] - centers[3][0], 2) + pow(c_3['col2'] - centers[3][1], 2))
c_4_dist = sum(pow(c_4['col1'] - centers[4][0], 2) + pow(c_4['col2'] - centers[4][1], 2))
```

```python
res = c_0_dist + c_1_dist + c_2_dist + c_3_dist + c_4_dist
res
```
