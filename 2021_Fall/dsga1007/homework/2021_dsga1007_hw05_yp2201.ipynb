**Problem 1.9 (Points 20)**

Normalize each of the three columns `math score`, `reading score`, `writing score` by subtracting its mean value and dividing by its standard deviation, and form three new columns: `math_score_norm`, `reading_score_norm`, `writing_score_norm`, respectively.

Perform the same task the one above (1.8) on these newly formed columns.  


```python
math_mean = np.mean(student_df['math score'])
math_std = np.std(student_df['math score'])

reading_mean = np.mean(student_df['reading score'])
reading_std = np.std(student_df['reading score'])

writing_mean = np.mean(student_df['writing score'])
writing_std = np.std(student_df['writing score'])


student_df['math_score_norm'] = (student_df['math score'] - math_mean) / math_std
student_df['reading_score_norm'] = (student_df['reading score'] - reading_mean) / reading_std
student_df['writing_score_norm'] = (student_df['writing score'] - writing_mean) / writing_std
```

```python
student_df.plot.scatter(x="math_score_norm", y="reading_score_norm", alpha=0.5)
plt.show()
```

```python
student_df.plot.scatter(x="math_score_norm", y="writing_score_norm", alpha=0.5)
plt.show()
```

```python
student_df.plot.scatter(x="reading_score_norm", y="writing_score_norm", alpha=0.5)
plt.show()
```
