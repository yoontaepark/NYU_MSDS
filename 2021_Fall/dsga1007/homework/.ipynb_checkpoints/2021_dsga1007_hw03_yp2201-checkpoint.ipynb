### Problem 02 (50 points)

In this problem we examine the beta distribution, characterized by two parameters $a$ and $b$, over all possible pairs $a, b \in \{0.5, 1, 2.0, 5.0\}$. Note that there are 16 such pairs.
  
More on the beta distribution one can find at: 
https://en.wikipedia.org/wiki/Beta_distribution


Your task is that for every pair of these 16 pairs of $(a, b)$:

1. (20 points) To generate a random vector obtained by the beta distribution $B(a, b)$ of size `n_size = 100_000`. You can use the following function: 
https://numpy.org/doc/stable/reference/random/generated/numpy.random.beta.html
    
2. (30 points) And for this random vector, to plot **as a subplot** the corresponding histogram with the number of bins `n_bins = 50`.



Important Note: your solution should produce only one figure with $4 \times 4$ subplots (i.e. 4 subplots in each row and 4 in each column).

Other Note: "100_000" is a convenience notation for "100000"
    

```python
# (20 points) To generate a random vector obtained by the beta distribution  ùêµ(ùëé,ùëè) of size n_size = 100_000

n_size = 100_000
res = []


a = {0.5, 1, 2.0, 5.0}
b = {0.5, 1, 2.0, 5.0}

for a1 in a:
    for b1 in b:
        res.append(np.random.beta(a1, b1, size=n_size))
        
print(res)
```

```python
# (30 points) And for this random vector, to plot as a subplot the corresponding histogram with the number of 
# bins n_bins = 50
n_bins = 50

fig, axs = plt.subplots(4, 4, figsize=(16,10))

axs[0][0].hist(res[0], bins=n_bins)
axs[0][1].hist(res[1], bins=n_bins)
axs[0][2].hist(res[2], bins=n_bins)
axs[0][3].hist(res[3], bins=n_bins)

axs[1][0].hist(res[4], bins=n_bins)
axs[1][1].hist(res[5], bins=n_bins)
axs[1][2].hist(res[6], bins=n_bins)
axs[1][3].hist(res[7], bins=n_bins)

axs[2][0].hist(res[8], bins=n_bins)
axs[2][1].hist(res[9], bins=n_bins)
axs[2][2].hist(res[10], bins=n_bins)
axs[2][3].hist(res[11], bins=n_bins)

axs[3][0].hist(res[12], bins=n_bins)
axs[3][1].hist(res[13], bins=n_bins)
axs[3][2].hist(res[14], bins=n_bins)
axs[3][3].hist(res[15], bins=n_bins)

plt.show()
```
