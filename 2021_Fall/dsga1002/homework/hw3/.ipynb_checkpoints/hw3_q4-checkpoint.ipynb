## Q4-a) Corresponding histogram shows a Uniform Distribution

```python
# 4-a)
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (10, 3)
plt.rcParams['font.size'] = 12

data = np.load('samples.npy')
param = 1.0
x = data
y = 1 - np.exp(-param * x)

plt.title('Corresponding histogram: Uniform Distribution, bins=20')
plt.hist(y, bins=20, edgecolor = 'black', density=True)

plt.show()
```
