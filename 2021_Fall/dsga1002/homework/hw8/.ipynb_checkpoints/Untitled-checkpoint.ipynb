### Q4 - a)
The moving average is plotted below. It converges to 3.5.  
If we assume that the expected value of x is finite then it is equal to zero,  
since f(x) is an even function and consequently xf(x) is odd so that E(x) = integral(xf(x)dx) = 0.  

By the Law of Large Numbers we should estimate c to equal 3.5 since a â†’ c in mean square and in probability.  
This works under the assumption that the x has finite mean and variance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
r1 = np.loadtxt('./radioactive_sample_1.txt')
r2 = np.loadtxt('./radioactive_sample_2.txt')

q4_a_list = []

for i in range(len(r1)):
    q4_a_list.append(np.mean(r1[:i+1]))

plt.plot(q4_a_list)
plt.show()    
```
