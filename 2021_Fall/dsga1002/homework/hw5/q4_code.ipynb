```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

cnt = 0
u1,u2 = [],[]

while cnt < 1000:
    x, y = np.random.rand(2)
    
    if y <= 2*x and y <= 2*(1-x):
        u1.append(x)
        u2.append(y)
        cnt += 1    
```

```python
plt.scatter(u1, u2)
plt.show()
```
